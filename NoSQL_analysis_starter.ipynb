### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene" : 20}

# Use count_documents to display the number of documents in the result
print("There are", db.establishments.count_documents(query), "restaraunts with a hygiene score of exactly 20")

# Display the first document in the results using pprint
pprint(db.establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
# not pd.DataFrame, have to use json_normalize
results = db.establishments.find(query)
df = pd.json_normalize(results)

# Display the number of rows in the DataFrame
print("There are", len(df), "rows in the dataframe")

# Display the first 10 rows of the DataFrame
df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {"LocalAuthorityName": {"$regex": "London"}, "RatingValue" : {"$gte" : 4}}

# Use count_documents to display the number of documents in the result
print("There are", db.establishments.count_documents(query), "restaraunts with a rating value of 4 or more")

# Display the first document in the results using pprint
pprint(db.establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
results = db.establishments.find(query)
dftwo = pd.json_normalize(results)

# Display the number of rows in the DataFrame
print("There are", len(dftwo), "rows in the dataframe")

# Display the first 10 rows of the DataFrame
dftwo.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# find Penang Flavors, code from other notebook
rez = db.establishments.find({"BusinessName" : "Penang Flavours"})

for i in rez:
    pprint(i)
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

query = {'geocode.latitude':
                {'$gte':latitude-degree_search, '$lte': latitude+degree_search},
         'geocode.longitude': 
                {'$gte':longitude-degree_search, '$lte': longitude+degree_search},
         'RatingValue': 5
        }
sort = [('score.Hygiene', 1)]
limit = 5

# Print the results
results = db.establishments.find(query).sort(sort).limit(limit)
for i in results:
    pprint(i)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Print the number of documents in the result
matching = {"$match": {"scores.Hygiene": 0}}
grouping = {"$group": 
                {'_id': "$LocalAuthorityName", "total": 
                    {"$sum": 1}}
           }
sorting = {"$sort": {"num_estabs": -1}}

pipeline = [matching, grouping, sorting]

results = list(db.establishments.aggregate(pipeline))

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
dffour = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print("There are", len(dffour), "rows in the dataframe")

# Display the first 10 rows of the DataFrame
dffour.head(10)
```
